```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
clv = df.groupby('Customer ID')['Total Spent'].sum().reset_index()
clv.columns = ['CustomerID','CLV']
clv.head()
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(clv[['CLV']])
kmeans=KMeans(n_clusters=3,random_state=42)
clv['Cluster']=kmeans.fit_predict(X)
clv.head()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=clv,x='CustomerID',y='CLV',hue='Cluster')
plt.title('Customer Segmentation by CLV')
plt.show()
```

```python
monthly_sales = df.groupby('Month')['Total Spent'].sum().reset_index()
monthly_sales.columns = ['Month','monthly_sales']
monthly_sales.head()
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(monthly_sales[['monthly_sales']])
kmeans=KMeans(n_clusters=3,random_state=42)
monthly_sales['Cluster']=kmeans.fit_predict(X)
monthly_sales.head()
```

```python
plt.figure(figsize=(12,7))
sns.barplot(data=monthly_sales, x='Month', y='monthly_sales', hue='Cluster')

plt.title("Monthly Sales grouped by Cluster", fontsize=16)
plt.xticks(rotation=45)
plt.show()
```
